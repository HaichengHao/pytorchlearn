# 多分类问题 

## softmax多分类问题

神经网络的原始输出不是一个概率值,实质上只是输入的数值  
做了复杂的加权和与分线性处理之后的一个值而已,那么如何将  
这个输出变为概率分布 ?  
这就是`Softmax`层的作用

softmax分类  
对数几率回归解决的是二分类问题，我们用的激活函数是sigmod  
对于多个选项的问题,我们可以使用softmax函数  
它是对数几率回归在N个可能不同的值上的推广

$$\sigma(z)_j = \frac{e^(z_j)}{\sum_{k=1}^K},for j= 1,...,K$$  
softmax要求每个样本必须属于某个类别,且所有可能的样本均被覆盖  
softmax样本分量之和为1  
当只有两个类别时,与对数几率回归完全相同   
在pytorch中,对于多酚类问题我们使用  
`nn.CrossEntropyLoss()`和`nn.NLLLoss`等来计算softmax的交叉熵

#### 将数据转换成张量 

````python
train_x = torch.from_numpy(x_train).type(torch.float32)
train_y = torch.from_numpy(y_train).type(torch.int64)
test_x = torch.from_numpy(x_test).type(torch.float32)
test_y = torch.from_numpy(y_test).type(torch.int64)
````

#### 使用Tensordataset和dataloader

````python
batch = 8 #定义batchsize
DS_train = TensorDataset(train_x,train_y)
DL_train = DataLoader(DS_train,batch_size=batch,shuffle=True)

DS_test = TensorDataset(test_x,test_y)
DS_test = DataLoader(DS_test,batch_size=batch)
````

#### 创建模型

````python
class Model(nn.Module):
    def __init__(self): 
        super().__init__() 
        self.liner_1 = nn.Linear(4,32)#第一层4个特征输入
        self.liner_2 = nn.Linear(32,32) 
        self.liner_3 = nn.Linear(32,3) #三分类问题,最后一层3
    def forward(self,input):
        x = F.relu(self.liner_1(input))
        x = F.relu(self.liner_2(x)) 
        # x = F.softmax(self.liner_3(x))  #这样是不行的,不进行激活 
        x = self.liner_3(x)
        return x 
````

### 创建训练函数

#### 创建获取准确率的函数

````python
def acc(y_pred,y_true): 
    y_pred = torch.argmax(y_pred,dim=1)
    acc = (y_pred == y_true).float().mean() 
    return acc
````

#### 定义获取模型和优化函数的方法

````python
lr = 0.0001
def get_model():
    model = Model() 
    optm = torch.optim.Adam(model.parameters(),lr=lr) 
    return model,optm
````

````python
model,optim = get_model()
````

#### 开训练

````python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
epochs = 20
# loss_fn = nn.BCELoss()
# 提前将验证用的数据放到GPU上
X_train = train_x.to(device)
Y_train = train_y.to(device)
X_test = test_x.to(device)
Y_test = test_y.to(device)

#创建俩列表接收损失值和精度值
train_loss_lst = []
train_acc_lst = []
tst_loss_lst=[]
tst_acc_lst=[]
for epoch in range(epochs):
    model.train()  # 设置为训练模式
    for x, y in DL_train:
        x = x.to(device)
        y = y.to(device)
        
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        
        optim.zero_grad()
        loss.backward()
        optim.step()

    # 每个epoch结束后评估模型
    model.eval()  # 设置为评估模式
    with torch.no_grad():
        y_pred = model(X_train)
        epoch_acc= acc(y_pred,Y_train).item() #加上item()将张量转换为第标量
        train_acc_lst.append(epoch_acc)
        loss_ = loss_fn(y_pred, Y_train).item()
        train_loss_lst.append(loss_)

        y_pred_tst = model(X_test)
        epochtest_acc= acc(y_pred_tst,Y_test).item() #加上item()将张量转换为第标量
        tst_acc_lst.append(epochtest_acc)
        
        losstest_ = loss_fn(y_pred_tst, Y_test).item()
        tst_loss_lst.append(losstest_)
        
        
        print(f'轮次 >> {epoch} | 损失: {loss_:.3f} | 轮次平均正确率:{epoch_acc:.3f}  \
        | 测试集损失:{losstest_:.3f} | 测试集准确率:{epochtest_acc:.3f}')

print('训练结束')
# print(loss_lst)
````

### 编写fit函数来复用自己的模型

编写fit  
输入模型,输入数据,对输入数据在模型上训练  
并且能够返回loss和acc

````python

def fit_cpu(Train_DL,TST_DL,Model,epoch_,optim,loss_fn): 
    correct = 0
    total = 0
    running_loss = 0
    tstcorrect = 0
    tsttotal = 0
    tstrunning_loss = 0


    for x, y in Train_DL:
        y_pred = Model(x)
        loss = loss_fn(y_pred, y)
        optim.zero_grad()
        loss.backward()
        optim.step()
        with torch.no_grad():
            y_pred = torch.argmax(y_pred,dim=1)
            correct += (y_pred == y).sum().item() #预测正确的个数
            # 因为(y_pred == y).sum()是一个张量,所以为了灯饰可以数值操作,我们用.item()取值
            total += y.size(0) #样本的个数
            running_loss += loss.item()

        
    epoch_acc = correct/total
    epoch_loss = running_loss/len(Train_DL.dataset)

    with torch.no_grad():
         for x, y in TST_DL:
            y_pred = Model(x)
            loss = loss_fn(y_pred, y)
            y_pred = torch.argmax(y_pred,dim=1)   #获取真正的预测结果,不懂就往上巴拉argmax
            tstcorrect += (y_pred == y).sum().item() #预测正确的个数
            tsttotal += y.size(0) #样本的个数
            tstrunning_loss += loss.item()
    tstepoch_acc = tstcorrect/tsttotal
    tstepoch_loss = tstrunning_loss/len(TST_DL.dataset)

    print(
        f'epoch:{epoch} | loss:{epoch_loss:.3f} | acc:{epoch_acc:.3f} | tstloss:{tstepoch_loss:.3f} | tstacc:{tstepoch_acc:.3f}'
    )

    return epoch_loss,epoch_acc,tstepoch_loss,tstepoch_acc
          
````

````python
model_cpu = Model()
optim_cpu = torch.optim.Adam(model_cpu.parameters(),lr=0.001)
loss_fn_cpu = nn.CrossEntropyLoss() 
batch = 8 #定义batchsize
DS_train = TensorDataset(train_x,train_y)
DL_train = DataLoader(DS_train,batch_size=batch,shuffle=True)

DS_test = TensorDataset(test_x,test_y)
DL_test = DataLoader(DS_test,batch_size=batch)  

epochs = 20 

train_loss_cpu=[] 
train_acc_cpu =[] 
tst_loss_cpu = [] 
tst_acc_cpu=[] 
for epoch in range(epochs): 
    epoch_loss,epoch_acc,tstepoch_loss,tstepoch_acc = fit_cpu(Train_DL=DL_train,TST_DL=DL_test,Model=model_cpu,epoch_=epoch,optim=optim_cpu,loss_fn=loss_fn_cpu)
    train_loss_cpu.append(epoch_loss)
    train_acc_cpu.append(epoch_acc)
    tst_loss_cpu.append(tstepoch_loss)
    tst_acc_cpu.append(tstepoch_acc)
````

````python
def fit_gpu(Train_DL,TST_DL,Model_m,epoch_,optim,loss_fn): 
    correct = 0
    total = 0
    running_loss = 0
    tstcorrect = 0
    tsttotal = 0
    tstrunning_loss = 0
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    Model_m.to(device)
    # loss_fn = nn.BCELoss()
    # 提前将验证用的数据放到GPU上
    
    # X_train = train_x.to(device)
    # Y_train = train_y.to(device)
    # X_test = test_x.to(device)
    # Y_test = test_y.to(device)
   
    #创建俩列表接收损失值和精度值
    train_loss_lst = []
    train_acc_lst = []
    tst_loss_lst=[]
    tst_acc_lst=[]
  
    Model_m.train()  # 设置为训练模式
    for x, y in Train_DL:
        x = x.to(device)
        y = y.to(device)
        
        y_pred = Model_m(x)
        loss = loss_fn(y_pred, y)
        
        optim.zero_grad()
        loss.backward()
        optim.step()

        # 每个epoch结束后评估模型
   
        with torch.no_grad():
            y_pred = torch.argmax(y_pred,dim=1)
            correct += (y_pred == y).sum().item() #预测正确的个数
            # 因为(y_pred == y).sum()是一个张量,所以为了灯饰可以数值操作,我们用.item()取值
            total += y.size(0) #样本的个数
            running_loss += loss.item()

        
    epoch_acc = correct/total
    epoch_loss = running_loss/len(Train_DL.dataset)

    #测试阶段 
    Model_m.eval()
    with torch.no_grad(): #测试不需要反向传播,而是用训练好的模型来测试测试集的数据
         for x, y in TST_DL:
             x, y = x.to(device), y.to(device)
             y_pred =Model_m(x)
             loss = loss_fn(y_pred, y)
             
             y_pred = torch.argmax(y_pred,dim=1)   #获取真正的预测结果,不懂就往上巴拉argmax
             tstcorrect += (y_pred == y).sum().item() #预测正确的个数
             tsttotal += y.size(0) #样本的个数
             tstrunning_loss += loss.item()
             
    tstepoch_acc = tstcorrect/tsttotal
    tstepoch_loss = tstrunning_loss/len(TST_DL.dataset)

    print(
        f'epoch:{epoch} | loss:{epoch_loss:.3f} | acc:{epoch_acc:.3f} | tstloss:{tstepoch_loss:.3f} | tstacc:{tstepoch_acc:.3f}'
    )

    return epoch_loss,epoch_acc,tstepoch_loss,tstepoch_acc
````

在 PyTorch 中，并没有官方提供的名为 `fit` 的通用函数用于训练模型。通常，PyTorch 用户需要手动编写训练循环来执行模型的训练过程。这包括定义如何前向传播、计算损失、反向传播以及更新权重等步骤。这种灵活性允许用户根据自己的需求定制训练过程，但同时也要求更多的代码和对深度学习流程的理解。

不过，虽然PyTorch本身没有提供这样的功能，但你可以使用或参考一些第三方库或工具，比如：

- **Ignite**：由 PyTorch 官方团队支持的一个高级库，简化了训练和评估的循环。
- **Lightning**：一个社区驱动的框架，旨在减少样板代码（boilerplate code），同时保持对研究的灵活性。

这些工具提供了更高层次的API，可以让你通过类似 `fit` 方法的方式来训练模型，大大简化了代码量。

例如，在 PyTorch Lightning 中，你可以创建一个继承自 `LightningModule` 的类，并实现 `training_step`, `validation_step` 等方法，然后通过调用 `trainer.fit(model)` 来训练你的模型。

### 示例代码 - 使用 PyTorch Lightning

```python
import pytorch_lightning as pl
import torch
from torch import nn
import torch.nn.functional as F

class MNISTModel(pl.LightningModule):
    def __init__(self):
        super(MNISTModel, self).__init__()
        self.layer = nn.Linear(28 * 28, 10)

    def forward(self, x):
        return torch.relu(self.layer(x.view(x.size(0), -1)))

    def training_step(self, batch, batch_idx):
        x, y = batch
        logits = self(x)
        loss = F.nll_loss(logits, y)
        return loss
    
    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=1e-3)

# trainer = pl.Trainer(max_epochs=5)
# trainer.fit(model, train_loader)
```

这段代码展示了一个简单的例子，说明了如何使用 PyTorch Lightning 来组织你的模型训练过程。通过这种方式，你可以利用类似于 `fit` 函数的功能，而不需要从头开始构建整个训练循环。如果你正在寻找一种简化训练流程的方法，考虑使用这些工具可能会对你有所帮助。
